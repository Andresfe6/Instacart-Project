### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

````python
# Contar las veces que cada producto aparece en los pedidos
product_counts = order_products.groupby('product_id').size()

# Calcular la tasa de repetición para cada producto
total_orders = len(order_products['order_id'].unique())  # Total de pedidos únicos
repeat_rate = product_counts / total_orders

# Obtener los 20 productos con la mayor tasa de repetición
top_20_repeat_rate = repeat_rate.sort_values(ascending=False).head(20)

# Unir con el DataFrame 'products' para obtener el nombre del producto
top_20_repeat_rate_info = products[products['product_id'].isin(top_20_repeat_rate.index)]

# Mostrar los 20 productos con la mayor tasa de repetición
top_20_repeat_rate_info = top_20_repeat_rate_info.set_index('product_id')
top_20_repeat_rate_info['repeat_rate'] = top_20_repeat_rate.values

print(top_20_repeat_rate_info[['product_name', 'repeat_rate']])
````

### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

````python
# Paso 1: Crear un DataFrame con las compras por cliente
user_products = order_products[['order_id', 'product_id']].merge(instacart_orders[['order_id', 'user_id']], on='order_id')

# Paso 2: Contar cuántas veces cada producto fue comprado por cada cliente
product_repeat_counts = user_products.groupby(['user_id', 'product_id']).size().reset_index(name='product_count')

# Paso 3: Filtrar los productos que se repiten más de una vez
repeated_products = product_repeat_counts[product_repeat_counts['product_count'] > 1]

# Paso 4: Calcular la tasa de repetición por cliente
# Contamos el total de productos que compró cada cliente
total_products_per_user = product_repeat_counts.groupby('user_id')['product_id'].nunique()

# Calcular la tasa de repetición
repeat_rate_per_user = repeated_products.groupby('user_id').size() / total_products_per_user

# Paso 5: Mostrar la tasa promedio de repetición por cliente
avg_repeat_rate = repeat_rate_per_user.mean()
print(f"La tasa promedio de repetición por cliente es: {avg_repeat_rate:.2f}")
````

Escribe aquí tus conclusiones

Una tasa demasaido baja, da para pensar que estamos haciendo mal para no retener a nuestros clientes
